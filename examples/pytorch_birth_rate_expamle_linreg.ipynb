```python
import torch
```

```python
import pandas as pd
from torch.autograd import Variable

import numpy as np
```

```python
data=pd.read_csv('birth_life_2010.txt',sep="\t",header=None)
data.columns=["a","b","c"]
```

```python
data1=data[1:]
```

```python
y=pd.Series(data1['c'])

type(torch.from_numpy(y.values.astype(float)))
```

```python
#y=Variable(torch.from_numpy(y.values).float())
y = Variable(torch.from_numpy(y.values.astype(float)))
```

```python
x=pd.Series(data1['b'])
x=Variable(torch.from_numpy(x.values.astype(float)))
```

```python
type(x)
```

```python
from torch.autograd import Variable
dtype=torch.cuda.FloatTensor
```

```python

w = Variable(torch.DoubleTensor(1,1),  requires_grad=True)

b=Variable(torch.DoubleTensor(1,1),  requires_grad=True)
```

```python
import torch.nn as nn
```

```python
criterion = nn.MSELoss()# Mean Squared Loss
l_rate = 0.001
optimiser = torch.optim.SGD([w,b], lr = l_rate) #Stochastic Gradient Descent

epochs = 2000
```

```python
w.data
```

```python

for t in range(epochs):
    y_pred=x*w+b
    loss=criterion(y_pred,y)
    
    print(t,loss.data[0])
    loss.backward()
    optimiser.step()
    optimiser.zero_grad()
```
